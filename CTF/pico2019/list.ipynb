```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
with open("q.html") as f:
    html = f.read()
soup = BeautifulSoup(html, 'lxml')
```

```python
l = []
for x in soup.find('div', id='problem-root-holder').find_all('span', class_='do-expand'):
    t = x.text
    l.append(t.split(" - ")[0].rstrip().replace(" ", "_"))
```

```python
l
```

```python
for name in l:
    if os.path.exists(name):
        continue
        
    os.mkdir(name)
```

```python
s = """
## {}
- 概要
    - mada
- 解法
    - mada

"""
```

```python
for name in l:
    print(s.format(name))
```
